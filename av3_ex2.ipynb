```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Carregar dados --------------------------------------------------------
T, M = np.loadtxt("/content/starts.txt", unpack=True)   # T ≃ 700-1500  ;  M ≃ 0-18

# Calc. densidade 2-D (kde)
xy = np.vstack([T, M])
densidade = gaussian_kde(xy)(xy)

fig, ax = plt.subplots(figsize=(8, 6))
sc = ax.scatter(T, M, c=densidade, s=8, cmap="viridis", alpha=0.6, lw=0)

# Ajustes de eixos
ax.invert_yaxis()                 # magnitudes menores = mais brilhantes
ax.set_xlabel("Temperatura (K)")
ax.set_ylabel("Magnitude")
ax.set_title("Diagrama H-R (densidade por cor)")
cbar = fig.colorbar(sc, ax=ax, label="Densidade relativa")
plt.tight_layout()
plt.show()
```
